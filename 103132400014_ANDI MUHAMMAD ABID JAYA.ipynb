```python
# Import library utama
import pandas as pd
import numpy as np

# Untuk imputasi dan encoding
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Membaca dataset yang telah diunduh dan diupload ke Colab
# Pastikan nama file sesuai yang ada di Drive
features = pd.read_csv('/features.csv')
labels = pd.read_csv('/labels.csv')

# Menampilkan 5 baris pertama dari masing-masing dataset
print("=== Features Dataset ===")
display(features.head())

print("=== Labels Dataset ===")
display(labels.head())
```

 Penjelasan kolom (contoh, sesuaikan dengan dataset yang sebenarnya)

id: ID unik untuk setiap entri

Age: Usia responden

Gender: Jenis kelamin responden

Income: Pendapatan per bulan

Target: Label hasil klasifikasi



```python
# Menggabungkan dua tabel berdasarkan kolom 'Student Id'
df = pd.merge(features, labels, left_on='Student Id', right_index=True)

print("=== Dataset Setelah Digabung ===")
display(df.head())
```

```python
print("=== Tipe Data Tiap Kolom ===")
print(df.dtypes)
```

```python
print("=== Statistik Deskriptif Dataset ===")
display(df.describe(include='all'))

# Analisis singkat
# Mean dan median bisa menggambarkan sebaran data
# Kolom kategorikal (object) nantinya perlu encoding
# Perhatikan nilai min, max, dan std dev untuk melihat sebaran
```

```python
# 20 baris pertama
print("=== 20 Baris Pertama ===")
display(df.head(20))

# 20 baris terakhir
print("=== 20 Baris Terakhir ===")
display(df.tail(20))

# Baris dengan index 80 - 100
print("=== Baris Index 80 - 100 ===")
display(df.iloc[80:101])
```

```python
# Cek jumlah missing value per kolom
print("=== Jumlah Nilai Hilang Sebelum Imputasi ===")
print(df.isnull().sum())
```

```python
# Contoh imputasi menggunakan mean untuk kolom numerik
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
imputer_mean = SimpleImputer(strategy='mean')

df[num_cols] = imputer_mean.fit_transform(df[num_cols])

print("=== Jumlah Nilai Hilang Setelah Imputasi ===")
print(df.isnull().sum())
```

```python
# Contoh penggunaan KNNImputer untuk kolom numerik
# Gunakan jika ingin nilai plus
# imputer_knn = KNNImputer(n_neighbors=3)
# df[num_cols] = imputer_knn.fit_transform(df[num_cols])
# print("=== Setelah KNN Imputation ===")
# print(df.isnull().sum())
```

```python
# Cek kolom bertipe object
cat_cols = df.select_dtypes(include=['object']).columns
print("Kolom Kategorikal:", cat_cols.tolist())
```

```python
# Gunakan LabelEncoder untuk kolom kategorikal
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print("=== Dataset Setelah Label Encoding ===")
df.head()
```

```python
print("=== Dataset Final Setelah Preprocessing ===")
display(df.head())
df.info()

# Pastikan tidak ada nilai NaN
print("Jumlah Nilai Hilang:", df.isnull().sum().sum())
```

```python
# Simpan hasil akhir ke file CSV
df.to_csv('/content/NIM_Nama_cleaned.csv', index=False)
print("Dataset final telah disimpan sebagai NIM_Nama_cleaned.csv")
```
