6. Tangani nilai yang hilang (NaN) menggunakan salah satu metode :


Dilakukan dengan df.isnull().sum().

Ditemukan beberapa kolom yang memiliki nilai kosong, terutama pada fitur ekonomi seperti tuition fees up to date dan scholarship holder.
Nilai-nilai ini nantinya akan diimputasi dengan median atau modus.

```python
# Mengecek jumlah nilai yang hilang di setiap kolom
print("=== Missing Values Sebelum Imputasi ===")
print(df.isnull().sum())
```

```python
# Memisahkan kolom numerik dan kategorikal
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Membuat objek imputasi
from sklearn.impute import SimpleImputer
```

```python
# Median untuk data numerik
imputer_num = SimpleImputer(strategy='median')
df[num_cols] = imputer_num.fit_transform(df[num_cols])
```

```python
# Mode untuk data kategorikal
imputer_cat = SimpleImputer(strategy='most_frequent')
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])
```

```python
# Mengecek kembali setelah imputasi
print("\n=== Jumlah Missing Values Setelah Imputasi ===")
print(df.isnull().sum().head(10))
```

7. Encoding terhadap kolom bertipe Object menjadi Numerik

Dilakukan label encoding untuk kolom kategorikal seperti gender dan marital status agar dapat digunakan dalam model numerik.

Hasilnya, setiap kategori direpresentasikan sebagai angka

```python
# Melakukan encoding pada kolom kategorikal
encoder = LabelEncoder()
for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])

print("=== Data Setelah Encoding (5 Baris Pertama) ===")
display(df.head())
```

```python
# Cek tipe data setelah encoding
print("\nTipe data setelah encoding:")
print(df.dtypes)
```
