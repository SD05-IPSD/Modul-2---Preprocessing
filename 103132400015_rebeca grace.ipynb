Analisis deskriptif dari dataset:

Output dari `df.describe()` memberikan ringkasan statistik untuk kolom-kolom numerik dalam dataset. Beberapa poin penting yang dapat diamati adalah:

*   **Count:** Menunjukkan jumlah nilai non-null untuk setiap kolom. Terlihat bahwa ada beberapa kolom dengan jumlah nilai yang kurang dari total baris (3096), menandakan adanya nilai yang hilang (NaN).
*   **Mean:** Rata-rata nilai untuk setiap kolom. Ini memberikan gambaran tentang nilai pusat dari distribusi data.
*   **Std:** Standar deviasi, mengukur sebaran data dari rata-rata. Nilai standar deviasi yang tinggi menunjukkan sebaran data yang lebih luas.
*   **Min dan Max:** Menunjukkan nilai minimum dan maksimum dalam setiap kolom. Ini membantu mengidentifikasi rentang nilai data.
*   **25%, 50%, 75% (Quartiles):** Menunjukkan kuartil pertama, median (kuartil kedua), dan kuartil ketiga. Ini memberikan wawasan tentang distribusi data dan keberadaan outlier.

Dari analisis ini, kita dapat melihat adanya variasi yang signifikan dalam beberapa kolom, serta keberadaan nilai yang hilang yang perlu ditangani lebih lanjut.

6. tangani nilai yang hilang (NaN) menggunakan salah satu metode berikut:

```python
#menampilkan data sebelum di imputasi
print(df.isnull().sum())
print("\ntotal nilai hilang sebelum imputasi: ", df.isnull().sum().sum())
```

```python
#ubah kategori menjadi angka (karna imputasi hanya bisa menggunakan angka)
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include=['object']):
    df_encoded[col] = df_encoded[col].astype('category').cat.codes
```

```python
#Buat dan jalankan SimpleImputer ---
# Pilihan strategi:
# ganti nan dengan rata - rata
imputer = SimpleImputer(strategy='mean')
```

```python
# Lakukan imputasi
df_imputed = pd.DataFrame(imputer.fit_transform(df_encoded),
                          columns=df_encoded.columns)
```

```python
#menampilkan jumlah data hilang sesudah imputasi
print(df_imputed.isnull().sum())
print("\ntotal nilai hilang sesudah imputasi: ",df_imputed.isnull().sum().sum())
```
