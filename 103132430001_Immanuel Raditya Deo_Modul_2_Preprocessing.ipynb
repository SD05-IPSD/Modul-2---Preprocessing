**Penjelasan Dataset (Dataset Summary)**



Dataset ini berisi informasi demografis, sosioekonomi,
dan akademik dari mahasiswa.

---
**Penjelasan setiap kolom:**
* **`Unnamed: 0`**: Nomor indeks baris data yang dibuat otomatis sebagai penunjuk setiap record data saja (bukan UID).
* **`Student Id`**: Nomor identifikasi unik yang diberikan kepada setiap mahasiswa.
* **`Marital status`**: Status perkawinan mahasiswa
* **`Application mode`**: Kode numerik yang mewakili jalur pendaftaran mahasiswa
* **`Application order`**: Urutan prioritas pilihan jurusan yang diambil. Angka yang lebih kecil berarti jurusan tersebut adalah pilihan utama.
* **`Course`**: Kode khusus untuk program studi spesifik yang diambil mahasiswa.
* **`Daytime/evening attendance`**: Menunjukkan waktu perkuliahan mahasiswa
* **`Previous qualification`**: Kode untuk tingkat pendidikan terakhir mahasiswa sebelum masuk universitas
* **`Previous qualification (grade)`**: Nilai akhir dari tingkat pendidikan sebelumnya.
* **`Nacionality`**: Kewarganegaraan mahasiswa.
* **`Mother's qualification`** & **`Father's qualification`**: Kode untuk tingkat pendidikan tertinggi yang dimiliki oleh ibu dan ayah.
* **`Mother's occupation`** & **`Father's occupation`**: Kode untuk jenis pekerjaan ibu dan ayah.
* **`Admission grade`**: Nilai akhir yang diperoleh mahasiswa saat seleksi masuk universitas.
* **`Displaced`**: Status perpindahan mahasiswa untuk kuliah apakah pergi dari daerah asal atau tidak
* **`Educational special needs`**: Indikator apakah mahasiswa memiliki kebutuhan pendidikan khusus.
* **`Debtor`**: Status tunggakan utang mahasiswa kepada universitas.
* **`Tuition fees up to date`**: Status pembayaran biaya kuliah.
* **`Gender`**: Jenis kelamin mahasiswa.
* **`Scholarship holder`**: Status sebagai penerima beasiswa.
* **`Age at enrollment`**: Usia mahasiswa saat pertama kali mendaftar.
* **`International`**: Status jika tergolong mahasiswa internasional atau tidak.
* **`Curricular units (credited)`**: Jumlah SKS/mata kuliah yang diakui dari pendidikan sebelumnya.
* **`Curricular units (enrolled)`**: Jumlah total SKS/mata kuliah yang diambil pada semester tersebut.
* **`Curricular units (evaluations)`**: Jumlah ujian/evaluasi yang diikuti oleh mahasiswa pada semester tersebut.
* **`Curricular units (approved)`**: Jumlah SKS/mata kuliah yang disetujui pada semester tersebut.
* **`Curricular units (grade)`**: Nilai rata-rata dari semua mata kuliah yang dievaluasi di semester itu.
* **`Curricular units (without evaluations)`**: Jumlah mata kuliah yang terdaftar namun tidak diikuti ujian/evaluasinya.
* **`Unemployment rate`**: Tingkat pengangguran di wilayah tersebut.
* **`Inflation rate`**: Tingkat inflasi di wilayah tersebut.
* **`GDP` (Gross Domestic Product)**: Produk Domestik Bruto, sebuah indikator pertumbuhan ekonomi wilayah/negara.
* **`Target`**: Variabel dependen yang menjadi tujuan prediksi nantinya dan untuk klasifikasi. Kolom ini berisi status akhir dari setiap mahasiswa (`Graduate`, `Dropout`, `Enrolled`)

###**DATA PREPROCESSING**

###6. **Handling missing values (*encoding & imputing*)**

```python
# Menampilkan dataframe terkini sebelum di encode and impute
df_mhs
```

```python
# Mengidentifikasi kolom yang memiliki null values
df_mhs.isna().sum()
```

```python
# Handling missing values with miceforest

# Pisahkan fitur dan target sebelum imputasi
X = df_mhs.drop('Target', axis=1)
y = df_mhs['Target']

# Membersihkan Nama Kolom Fitur
X.columns = [re.sub(r'[^A-Za-z0-9_]+', '_', col) for col in X.columns]

# Mengubah kolom yang bertipe data object menjadi category agar bisa dilakukan imputasi
object_cols = X.select_dtypes(include=['object']).columns

# Lakukan loop dan ubah tipe datanya
for col in object_cols:
    X[col] = X[col].astype('category')

# MICEforest akan menggunakan semua kolom di X untuk saling memprediksi nilai hilang
mhsk = mf.ImputationKernel(
  data=X,
  variable_schema=X.columns.to_list(),
  random_state=42,
)

# Jalankan proses imputasi
mhsk.mice(iterations=5, verbose=True)

# Dapatkan dataframe fitur yang sudah terisi penuh (imputed)
df_mhs_imputed = mhsk.complete_data()

print("\nDataframe Fitur Setelah Imputasi: ")
print(df_mhs_imputed)
print("\n")
df_mhs_imputed.info()
print("\n")
```

```python
# Menampilkan dataset setelah diimputasi
df_mhs_imputed
```

```python
# Agregasi lagi karena saat diimputasi tadi dipisahkan kolom target nya
df_mhs_new = pd.concat([df_mhs_imputed, train_labels], axis = 1)

df_mhs_new.info()
```

```python
# Encoding section (Menggunakan label encoder untuk kolom target saja, dan one hot encoder untuk beberapa kolom kategorikal yang independen)

# LABEL ENCODING
# Kolom yang akan diubah menggunakan One-Hot Encoding
ohe_cols = ['Marital_status', 'Daytime_evening_attendance_', 'Nacionality']
# Kolom yang akan diubah menggunakan Label Encoding
le_col = 'Target'

# Mengubah 'Graduate', 'Dropout', 'Enrolled' menjadi 0, 1, 2
le = LabelEncoder()

# Menerapkan fit_transform pada kolom target
df_mhs_new[le_col] = le.fit_transform(df_mhs_new[le_col])

# Membuat kamus penunjuk value pada kolom yang diubah menjadi angka apa
print(f"Mapping untuk kolom '{le_col}':")
for i, class_name in enumerate(le.classes_):
    print(f"  '{class_name}' -> {i}")
print("\n")

#ONE HOT ENCODER
# Menerapkan one hot encoder untuk beberapa kolom kategorikal
print("Menerapkan One-Hot Encoder pada Fitur")
df_mhs_encoded = pd.get_dummies(df_mhs_new, columns=ohe_cols,
                                prefix=ohe_cols, # Memberi nama kolom baru yang jelas
                                dummy_na=False,
                                dtype=int)

print("Dataframe Akhir (Setelah Encoding): ")
print(df_mhs_encoded.head())
print("\nTipe Data Kolom Setelah Encoding: ")
print(df_mhs_encoded.info())
```
