# 6. Tangani nilai yang hilang (NaN) menggunakan salah satu metode berikut:

```python
# Mengecek jumlah nilai hilang di setiap kolom
print("=== Jumlah Nilai Hilang per Kolom ===")
print(df.isnull().sum())

# Menghitung total keseluruhan nilai hilang
print("\nTotal nilai hilang di seluruh dataset:", df.isnull().sum().sum())

# Menghapus baris yang seluruh kolomnya kosong (jika ada)
df = df.dropna(how='all')

# Mengisi nilai hilang dengan rata-rata pada kolom numerik
df = df.fillna(df.mean(numeric_only=True))

# Mengecek ulang apakah masih ada nilai hilang
print("\n=== Setelah Penanganan ===")
print(df.isnull().sum())
```

# 7. Lakukan encoding terhadap kolom bertipe object menjadi numerik menggunakan:

```python
# Menampilkan kolom dengan tipe data object (kategori)
print("=== Kolom Kategorikal ===")
kategori_cols = df.select_dtypes(include=['object']).columns
print(kategori_cols)

# Melakukan encoding one-hot untuk kolom kategorikal
df_encoded = pd.get_dummies(df, columns=kategori_cols, drop_first=True)

# Menampilkan hasil encoding
print("\nData setelah encoding:")
display(df_encoded.head())

# Mengecek ukuran data setelah encoding
print("\nUkuran data setelah encoding:", df_encoded.shape)
```

# 8. Tampilkan kembali dataset yang telah melalui seluruh proses preprocessing (cleaning, imputasi, dan encoding).


```python
from sklearn.preprocessing import StandardScaler

# Menyalin dataset hasil encoding agar aman
df_scaled = df_encoded.copy()

# Memilih kolom numerik
num_cols = df_scaled.select_dtypes(include=['int64', 'float64']).columns

# Membuat objek scaler
scaler = StandardScaler()

# Melakukan standarisasi pada kolom numerik
df_scaled[num_cols] = scaler.fit_transform(df_scaled[num_cols])

# Menampilkan hasil normalisasi
print("=== Data Setelah Normalisasi ===")
display(df_scaled.head())

# Mengecek ukuran dataset akhir
print("\nUkuran data setelah normalisasi:", df_scaled.shape)
```
