# 6️. Tangani nilai yang hilang (NaN) menggunakan salah satu metode berikut:

## cek data yang hilang

```python
data.isnull().sum()
```

## Metode 1: Imputasi dengan Statistik (Mean, Median, Mode)

```python
# Contoh: imputasi dengan mean untuk kolom numerik
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
imputer_mean = SimpleImputer(strategy='mean')
data[num_cols] = imputer_mean.fit_transform(data[num_cols])

# Contoh: imputasi dengan mode untuk kolom kategorikal
cat_cols = data.select_dtypes(include=['object']).columns
imputer_mode = SimpleImputer(strategy='most_frequent')
data[cat_cols] = imputer_mode.fit_transform(data[cat_cols])

# Cek hasil setelah imputasi
data.isnull().sum()
```
