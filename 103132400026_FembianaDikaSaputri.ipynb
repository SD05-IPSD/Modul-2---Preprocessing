2. Melakukan agregasi antara tabel features dengan tabel labels

```python
# Menggabungkan kedua tabel berdasarkan index
merged = pd.concat([features, labels], axis=1)

# Menampilkan beberapa baris awal hasil penggabungan
print("\n=== Dataset Setelah Digabungkan ===")
display(merged.head())
```


---


- Proses agregasi dilakukan dengan menggabungkan (merge) tabel features dan labels
berdasarkan urutan baris yang sama (index).
- Hasilnya adalah dataset baru bernama "merged" yang berisi semua kolom fitur pelanggan serta kolom target (jumlah_promosi).
- Dataset hasil penggabungan ini akan digunakan untuk langkah-langkah analisis dan preprocessing berikutnya.

6. Menangani nilai yang hilang (NaN)

```python
# Mengecek jumlah nilai NaN sebelum imputasi
print("\n=== Jumlah Nilai NaN Sebelum Imputasi ===")
print(merged.isna().sum())
```

```python
# Menggunakan SimpleImputer (mean) untuk kolom numerik
num_cols = merged.select_dtypes(include=[np.number]).columns
imputer = SimpleImputer(strategy='mean')
merged[num_cols] = imputer.fit_transform(merged[num_cols])
```

Penjelasan


---

SimpleImputer melakukan imputasi otomatis berdasarkan strategi tertentu (mean, median, most_frequent, atau constant). Cocok digunakan untuk preprocessing dalam pipeline machine learning karena bentuknya sebagai object transformer.

- select_dtypes(include=[np.number]) → memilih semua kolom bertipe numerik (int atau float).

- SimpleImputer(strategy='mean') → membuat objek imputasi dari library sklearn untuk mengganti nilai yang hilang (NaN) dengan rata-rata (mean) dari setiap kolom.

- fit_transform() → dua proses dalam satu langkah:

- fit() menghitung nilai mean dari setiap kolom.

- transform() mengganti semua nilai NaN dengan mean yang sudah dihitung.

- merged[num_cols] = ... → hasil imputasi dimasukkan kembali ke kolom numerik di dataset merged.

---



```python
# Mengecek kembali setelah imputasi
print("\n=== Jumlah Nilai NaN Setelah Imputasi ===")
print(merged.isna().sum())
```

7. Encoding terhadap kolom bertipe object menjadi numerik

```python
# Menemukan kolom bertipe object (kategorikal)
cat_cols = merged.select_dtypes(include=['object']).columns
print("\n=== Kolom Kategorikal ===")
print(cat_cols)
```

```python
# Menggunakan LabelEncoder
encoder = LabelEncoder()
for col in cat_cols:
    merged[col] = encoder.fit_transform(merged[col].astype(str))

print("\n=== Dataset Setelah Encoding ===")
display(merged)
```

8. Menampilkan kembali dataset hasil preprocessing

```python
print("\n=== Dataset Setelah Semua Tahapan Preprocessing ===")
display(merged)

# Mengecek apakah masih ada nilai NaN atau tipe data tidak sesuai
print("\n=== Cek Nilai NaN Setelah Semua Tahapan ===")
print(merged.isna().sum())

print("\n=== Cek Tipe Data Akhir ===")
print(merged.dtypes)
```
