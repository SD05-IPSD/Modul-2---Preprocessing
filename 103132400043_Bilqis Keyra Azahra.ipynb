```python
print("\nJumlah nilai hilang sebelum imputasi:")
print(df.isnull().sum())
```

Perintah tersebut digunakan untuk menampilkan jumlah nilai yang hilang di setiap kolom sebelum dilakukan proses imputasi. Baris pertama hanya menampilkan teks penjelas di ouput, sedangkan baris kedua menghitung dan menampilkan brp banyak nilai nan di setiap kolom pd dtaframe df. dengan hasil ini dapat mengetahui kolom mana saja yang masih memiliki data kosong dan perly dibersihkan sblm analisis atau pemodelan dilakukan.

```python
# --- metode imputasi ---
from sklearn.impute import SimpleImputer

# imputasi dengan mean
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=[np.number])),
                          columns=df.select_dtypes(include=[np.number]).columns)

# Gabungkan kembali kolom non-numerik
non_numeric = df.select_dtypes(exclude=[np.number]).reset_index(drop=True)
df_imputed = pd.concat([df_imputed, non_numeric], axis=1)

print("\nJumlah nilai hilang setelah imputasi:")
print(df_imputed.isnull().sum())
```

Diatas menunjukkan proses imputasi data dengan mengganti nilai kosong pada kolom numerik menggunakan rata-rata melalui simpleimputer dari library scikit-learn. Setelah kolom non numerik digabung kembali agar struktur dataset tetap lengkap. Langkah ini memastikan untuk seluruh kolom numerik pd dataframe tidak memiliki nilai hilang, sehingga datset menjadi bersih dan siap digunakan.

```python
print("\n=== Dataset Final setelah Preprocessing (Cleaning, Imputasi, Encoding) ===")
display(df_imputed.head())

print("\nProses preprocessing selesai ")
```
