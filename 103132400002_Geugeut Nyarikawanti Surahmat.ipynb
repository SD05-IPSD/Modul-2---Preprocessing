Proses imputasi dasar menggunakan metode most frequence (modus) untuk mengatasi nilai NaN pada data kategorik

```python
from sklearn.impute import SimpleImputer

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

num_imputer = SimpleImputer(strategy = 'median')
df_num = num_imputer.fit_transform(df[numeric_columns])

cat_imputer = SimpleImputer(strategy = 'most_frequent')
df_cat = cat_imputer.fit_transform(df[categorical_columns])
```

Membersihkan nama kolom dari karakter spesial

```python
features.columns = (
    features.columns
    .str.replace('[^A-Za-z0-9_]+', '_', regex=True)
    .str.strip('_')                                  
)
```

Encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in features.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    features[col] = le.fit_transform(features[col].astype(str))
    label_encoders[col] = le
```

Imputasi pakai randomforest

```python
import miceforest as mf
kernel = mf.ImputationKernel(
    data = features,
    save_all_iterations_data = True,
    random_state = 42
)

kernel.mice(3)
features_imputed = kernel.complete_data()
```

Hasil setelah dilakukan imputasi
