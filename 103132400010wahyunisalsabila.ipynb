Mennagani missing value dengan SimpleImputer

```python
imputer = SimpleImputer(strategy='median')
```

```python
kolom_numerik = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
df[kolom_numerik] = imputer.fit_transform(df[kolom_numerik])
#lakukann imputasi
```

```python
df.isnull().sum()
```

Karna masih ada missing value, kolom yg bertipe numerik masih jadi missing value, maka dr itu coba terapkan ke kolom katgorikal


```python
kolom_kategori = df.select_dtypes(include=['object']).columns
```

```python
imputer_kategori = SimpleImputer(strategy='most_frequent')
```

```python
df[kolom_kategori] = imputer_kategori.fit_transform(df[kolom_kategori])
```

```python
df.isnull().sum()
```

**Lakukan encoding**

```python
#cek type data dlu
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
kolom_encode = ['Target', 'Daytime/evening attendance\t']
```

```python
for kolom in kolom_encode:
    df[kolom] = encoder.fit_transform(df[kolom])
```
