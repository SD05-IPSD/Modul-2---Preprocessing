# Menyiapkan data

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from IPython.display import display

# atur opsi agar semua kolom terlihat
pd.set_option('display.max_columns', None)

# memuat dataset
df_features = pd.read_csv('features.csv')
df_labels = pd.read_csv('labels.csv')
```

# 1Ô∏è. Tampilkan data yang telah diunduh, kemudian jelaskan semua kolom yang terdapat pada dataset tersebut.

```python
print("--- 1. Data Awal ---")
print("\nData Features:")
display(df_features.head())

print("\nData Labels:")
display(df_labels.head())
```

# 2Ô∏è. Lakukan agregasi antara tabel features dengan tabel labels.

```python
print("--- 2. Agregasi (Penggabungan) Data ---")

# gabungkan kedua tabel dan menghapus kolom yang tidak perlu
df = pd.concat([df_features, df_labels], axis=1)
df = df.drop(columns=['Student Id', 'Unnamed: 0'])

# tampilkan data yang telah digabung
df.head()
```

# 3Ô∏è. Tampilkan tipe data dari setiap kolom yang terdapat dalam dataset.

```python
print(" 3. Tipe Data Setiap Kolom ")

df.info()
```

```python
print(" 4. Statistik Deskriptif ")

df.describe(include='all')
```

# 6Ô∏è. Tangani nilai yang hilang (NaN) menggunakan salah satu metode

```python
print(" 6. Penanganan Nilai Hilang (Imputasi) ")
print("\nJumlah nilai hilang SEBELUM imputasi:")
print(df.isnull().sum())

# Proses Imputasi
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
df[numerical_cols] = SimpleImputer(strategy='median').fit_transform(df[numerical_cols])
df[categorical_cols] = SimpleImputer(strategy='most_frequent').fit_transform(df[categorical_cols])

print("\nJumlah nilai hilang SETELAH imputasi:")
print(df.isnull().sum())
```

# 7Ô∏è. Opsional : Lakukan encoding terhadap kolom bertipe object menjadi numerik menggunakan:


*   LabelEncoder, atau
*   OneHotEncoder





```python
print(" 7. Encoding Kolom Kategorikal ")

df_encoded = df.copy()
le = LabelEncoder()

for col in categorical_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Menampilkan hasil setelah encoding
df_encoded.head()
```

# 8Ô∏è. Tampilkan kembali dataset yang telah melalui seluruh proses preprocessing (cleaning, imputasi, dan encoding).

```python
print(" 8. Dataset Final Hasil Preprocessing ")

df_final = df_encoded

# Menampilkan hasil akhir
df_final.head()
```
