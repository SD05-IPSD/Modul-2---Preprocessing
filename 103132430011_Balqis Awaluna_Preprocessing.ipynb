### 1. Import Library dan Load Dataset
* Mengimpor library utama yang digunakan untuk manipulasi dan analisis data.
* Membaca dua file CSV yang telah diunggah.
* Menampilkan jumlah kolom, tipe data, dan 5 baris pertama dari masing-masing dataset untuk memastikan data berhasil dimuat dengan benar.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Load dataset
features = pd.read_csv("features.csv")
labels = pd.read_csv("labels.csv")

# Tampilkan 5 baris pertama
display(features.head())
display(labels.head())
```

### 2. Jelaskan Kolom Dataset
Dataset features memiliki 38 kolom yang mencakup data mahasiswa, nilai akademik, serta indikator ekonomi. Sedangkan labels memiliki kolom Target berisi hasil akhir mahasiswa (Graduate/Dropout).

```python
# Melihat daftar kolom dan jumlahnya
print("Jumlah kolom pada features:", len(features.columns))
print("Nama kolom pada features:")
print(features.columns.tolist())

print("\nJumlah kolom pada labels:", len(labels.columns))
print("Nama kolom pada labels:")
print(labels.columns.tolist())
```

```python
# Melihat tipe data setiap kolom
print("\nTipe data pada features:")
print(features.dtypes)

print("\nTipe data pada labels:")
print(labels.dtypes)
```

```python
# Melihat beberapa baris pertama dari tiap data
print("\nContoh data pada features:")
display(features.head())

print("\nContoh data pada labels:")
display(labels.head())
```

### 3. Agregasi Dataset
Menggabungkan tabel features dan labels secara horizontal (berdasarkan kolom) menjadi satu dataset utuh bernama *df*.

### 6. Missing Values
Mengecek jumlah nilai hilang (NaN) pada setiap kolom. Lalu, menggunakan *SimpleImputer* untuk mengganti nilai kosong:
* Kolom numerik akan diisi dengan median.
* Kolom kategorikal akan diisi dengan modus (most frequent value).



```python
# Cek jumlah missing sebelum imputasi
print("Missing values sebelum imputasi:")
print(df.isnull().sum().head(10))
```

```python
# Pisahkan kolom numerik dan kategorikal
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Imputasi numerik dengan median
num_imputer = SimpleImputer(strategy='median')
df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])
```

```python
# Imputasi kategorikal dengan modus
cat_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])

print("Missing values setelah imputasi:")
print(df.isnull().sum().head(10))
```

### 7. Encoding Kolom Kategorikal
Mengubah kolom bertipe teks (object) menjadi angka agar bisa digunakan dalam analisis atau model machine learning.

```python
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

df.head()
```
