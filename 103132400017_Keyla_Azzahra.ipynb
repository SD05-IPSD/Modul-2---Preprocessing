```python
# Import library
import pandas as pd

# Membaca dataset
features = pd.read_csv('/content/features.csv')
labels = pd.read_csv('/content/labels.csv')

# Menampilkan nama kolom di masing-masing tabel
print("Kolom pada FEATURES:")
print(features.columns.tolist())

print("\nKolom pada LABELS:")
print(labels.columns.tolist())

# Mencari kolom yang sama di kedua tabel untuk dijadikan kunci penggabungan
common_cols = list(set(features.columns) & set(labels.columns))
print("\nKolom yang sama di kedua tabel:", common_cols)

# Melakukan agregasi (penggabungan) berdasarkan kolom yang sama
# Jika tidak ada kolom yang sama, penggabungan dilakukan secara horizontal (tanpa kunci)
if len(common_cols) > 0:
    merged_data = pd.merge(features, labels, on=common_cols)
else:
    merged_data = pd.concat([features, labels], axis=1)

# Menampilkan hasil penggabungan
print("\n=== Data Setelah Agregasi ===")
display(merged_data.head())

# Menampilkan ukuran dataset hasil agregasi
print("\nUkuran dataset setelah agregasi:", merged_data.shape)
```

Pada tahap ini, dilakukan proses penyatuan data dari dua sumber, yaitu tabel features dan labels. Prosedur dimulai dengan memuat kedua dataset, kemudian program secara cerdas mengidentifikasi kolom yang berfungsi sebagai kunci penghubung (kolom dengan nama yang identik di kedua tabel). Jika kunci penghubung ini terdeteksi, fungsi pd.merge() dieksekusi untuk melakukan penggabungan data secara relasional. Namun, jika tidak ditemukan kolom yang sama, program akan menggunakan strategi alternatif dengan menggabungkan kedua tabel secara berdampingan (horizontal) menggunakan pd.concat().

Untuk mendapatkan wawasan dari kolom non-numerik, metode ini dipanggil kembali dengan argumen include=['object'], yang menghasilkan statistik untuk data kategorikal seperti jumlah nilai unik dan modus (top).

Interpretasi dari output statistik ini mengungkapkan karakteristik distribusi data. Untuk data numerik, standar deviasi yang tinggi atau selisih yang signifikan antara nilai minimum dan maksimum mengisyaratkan sebaran data yang luas atau potensi adanya nilai ekstrem.
```python
# Import library
import pandas as pd
from sklearn.impute import SimpleImputer

# Membaca dan menggabungkan dataset
features = pd.read_csv('/content/features.csv')
labels = pd.read_csv('/content/labels.csv')

# Gabungkan data berdasarkan kolom yang sama atau secara horizontal
common_cols = list(set(features.columns) & set(labels.columns))
if len(common_cols) > 0:
    merged_data = pd.merge(features, labels, on=common_cols)
else:
    merged_data = pd.concat([features, labels], axis=1)

# Menampilkan data sebelum imputasi
print("=== Data Sebelum Imputasi (cek nilai NaN) ===")
display(merged_data.isnull().sum())

# Membuat objek SimpleImputer (strategi mean)
imputer = SimpleImputer(strategy='mean')

# Terapkan imputasi hanya pada kolom numerik
numeric_cols = merged_data.select_dtypes(include=['float64', 'int64']).columns
merged_data[numeric_cols] = imputer.fit_transform(merged_data[numeric_cols])

# Menampilkan data setelah imputasi
print("\n=== Data Setelah Imputasi (cek nilai NaN) ===")
display(merged_data.isnull().sum())

# Menampilkan sebagian data untuk melihat hasilnya
print("\n=== Cuplikan Data Setelah Imputasi ===")
display(merged_data.head(10))
```

Pada tahap ini, dilakukan proses penanganan nilai yang hilang (missing values), yang diidentifikasi sebagai NaN. Langkah-langkahnya adalah sebagai berikut:

1. Pertama, dataset features dan labels disatukan menjadi satu tabel data.

2. Jumlah nilai NaN di setiap kolom dihitung terlebih dahulu menggunakan .isnull().sum() untuk mengetahui kondisi awal data.

3. Selanjutnya, sebuah strategi imputasi disiapkan menggunakan SimpleImputer dengan metode 'mean', yang bertujuan untuk mengisi setiap nilai kosong dengan nilai rata-rata dari kolomnya masing-masing.

4. Penting untuk dicatat bahwa strategi ini hanya diterapkan pada kolom-kolom numerik untuk menghindari error dan menjaga integritas data kategorikal.

5. Setelah proses imputasi selesai, dilakukan verifikasi ulang untuk memastikan bahwa sudah tidak ada lagi nilai yang hilang di dalam dataset.

Melalui metode ini, semua nilai kosong pada kolom numerik berhasil diisi, menghasilkan dataset yang utuh dan siap untuk tahap analisis atau pemodelan machine learning lebih lanjut tanpa terkendala oleh missing values.

```python
# Import library
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Membaca dan menggabungkan dataset
features = pd.read_csv('/content/features.csv')
labels = pd.read_csv('/content/labels.csv')

# Gabungkan dataset berdasarkan kolom yang sama
common_cols = list(set(features.columns) & set(labels.columns))
if len(common_cols) > 0:
    merged_data = pd.merge(features, labels, on=common_cols)
else:
    merged_data = pd.concat([features, labels], axis=1)

# Menampilkan tipe data awal
print("=== Tipe Data Sebelum Encoding ===")
print(merged_data.dtypes)

# Menentukan kolom dengan tipe data object (kategori)
categorical_cols = merged_data.select_dtypes(include=['object']).columns
print("\nKolom kategorikal yang akan di-encode:", list(categorical_cols))

# Melakukan encoding menggunakan LabelEncoder
le = LabelEncoder()
for col in categorical_cols:
    merged_data[col] = le.fit_transform(merged_data[col].astype(str))

# Menampilkan hasil setelah encoding
print("\n=== Dataset Setelah Encoding ===")
display(merged_data.head())

# Menampilkan tipe data baru
print("\n=== Tipe Data Setelah Encoding ===")
print(merged_data.dtypes)
```

Pada tahap ini, dilakukan transformasi data kategorikal menjadi format numerik agar dapat diolah oleh algoritma machine learning. Prosesnya adalah sebagai berikut:

1. Pertama, dataset features dan labels disatukan.

2. Selanjutnya, program secara otomatis mengidentifikasi semua kolom yang berisi data non-numerik (tipe object).

3. Setiap kolom yang teridentifikasi kemudian diproses menggunakan LabelEncoder. Alat ini bekerja dengan memberikan sebuah "label" angka unik (misalnya, 0, 1, 2, ...) untuk setiap kategori teks yang berbeda di dalam kolom tersebut.

4. Terakhir, dataset hasil transformasi dan tipe datanya ditampilkan kembali untuk memverifikasi bahwa semua kolom telah berhasil diubah menjadi format numerik.

Soal Nomor 8

```python
# Import library
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Membaca dataset
features = pd.read_csv('/content/features.csv')
labels = pd.read_csv('/content/labels.csv')

# Gabungkan dataset (cleaning)
common_cols = list(set(features.columns) & set(labels.columns))
if len(common_cols) > 0:
    merged_data = pd.merge(features, labels, on=common_cols)
else:
    merged_data = pd.concat([features, labels], axis=1)

# Imputasi nilai hilang (NaN) menggunakan mean
imputer = SimpleImputer(strategy='mean')
numeric_cols = merged_data.select_dtypes(include=['float64', 'int64']).columns
merged_data[numeric_cols] = imputer.fit_transform(merged_data[numeric_cols])

# Encoding kolom kategorikal menggunakan LabelEncoder
categorical_cols = merged_data.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in categorical_cols:
    merged_data[col] = le.fit_transform(merged_data[col].astype(str))

# Menampilkan dataset akhir setelah semua preprocessing
print("=== Dataset Akhir Setelah Preprocessing Lengkap ===")
display(merged_data.head(20))

# Mengecek ulang apakah masih ada nilai NaN
print("\n=== Cek Nilai NaN Setelah Semua Tahapan ===")
print(merged_data.isnull().sum())
```

Tahapan ini menampilkan wujud final dari dataset yang telah melewati seluruh rangkaian prapemrosesan. Secara berurutan, proses yang telah dilakukan meliputi:

1. Penggabungan Data: Menyatukan informasi dari features dan labels ke dalam satu tabel yang koheren.

2. Imputasi Nilai Hilang: Mengisi semua sel data yang kosong (NaN) pada kolom numerik dengan nilai rata-rata (mean) dari masing-masing kolom.

3. Encoding Kategorikal: Mentransformasikan semua data yang bersifat tekstual menjadi representasi numerik menggunakan LabelEncoder.

Dataset yang ditampilkan pada output ini merupakan hasil akhir yang sudah bersih, lengkap, dan seragam secara tipe data, sehingga sepenuhnya siap untuk digunakan pada tahap pemodelan machine learning.
